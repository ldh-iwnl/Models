```python
import matplotlib
np.random.seed(0)
x=np.arange(5)
y=np.random.randn(5)
plt.style.use('ggplot')
fig,axes = plt.subplots(ncols=2)
matplotlib.use('nbagg')
#disable xkcd mode
plt.rcdefaults()
v_bars = axes[0].bar(x,y, color='red')
h_bars = axes[1].barh(x,y, color='green')
plt.show()
```

```python
fig,axes = plt.subplots()
x=np.arange(5)
y=np.random.randint(-5,5,5)
v_bars = axes.bar(x,y,color='lightblue')

for bar,height in zip(v_bars,y):
    if height < 0:
        bar.set(color='darkred', linewidth=0)
    else:
        bar.set(color='darkblue', linewidth=0)
```

```python
x=np.linspace(0,10,200)
y1=2*x+1
y2=3*x+1.2
y_mean = 0.5*x*np.cos(2*x)+2.5*x+1.1
fig,ax = plt.subplots()
ax.fill_between(x,y1,y2, color='lightblue')
ax.plot(x,y_mean, color='red')
```

```python
mean_values = [1,2,3]
variance = [0.2,0.4,0.6]
bar_label = ['bar1','bar2','bar3']
x_pos = list(range(len(bar_label)))
max_y = max(zip(mean_values,variance))
plt.ylim(0,(max_y[0]+max_y[1])*1.2)
plt.ylabel('variable y')
plt.xticks(x_pos, bar_label)
plt.bar(x_pos, mean_values, yerr=variance, alpha=0.3)
```

```python
z1 = np.array([1,2,3])
z2 = np.array([4,5,6])
bar_labels = ['bar1','bar2','bar3']
fig = plt.figure(figsize=(10,8))
y_pos = np.arange(len(z1))
y_pos = [x for x in y_pos]
plt.barh(y_pos,z1, color='g', alpha=0.5)
plt.barh(y_pos,-z2, color='b', alpha=0.5)
plt.xlim(-max(z2)-1,max(z1)+1)
plt.ylim(-1,len(z1))
plt.show()
```

```python
green_data = [1,2,3]
blue_data = [4,5,6]
red_date = [7,8,9]
bar_labels = ['bar1','bar2','bar3']

pos= list(range(len(green_data)))
width = 0.2
fig,ax = plt.subplots(figsize=(10,8))

plt.bar(pos, green_data, width, alpha=0.5, color='g')
plt.bar([p+width for p in pos], blue_data, width, alpha=0.5, color='b')
plt.bar([p+width*2 for p in pos], red_date, width, alpha=0.5, color='r')
```

```python
data = range(200,225,5)
bar_labels = ['bar1','bar2','bar3','bar4','bar5']
fig = plt.figure(figsize=(8,4))
y_pos = np.arange(len(data))
plt.yticks(y_pos,bar_labels)

bars = plt.barh(y_pos,data, color='g', alpha=0.5)
plt.vlines(min(data), -1,len(data), linestyles='dashed')
for b,d in zip(bars,data):
    plt.text(b.get_width()+5, b.get_y()+b.get_height()/2, str(d), ha='center', va='center')
plt.show()
```

```python
mean_values = range(10,18)
x_pos = list(range(len(mean_values)))

import matplotlib.colors as col
import matplotlib.cm as cm
cmap1 = cm.ScalarMappable(col.Normalize(min(mean_values),max(mean_values),cm.hot))
cmap2 = cm.ScalarMappable(col.Normalize(0,20,cm.cool))
plt.subplot(1,2,1)
plt.bar(x_pos, mean_values, color=cmap1.to_rgba(mean_values))
plt.subplot(1,2,2)
plt.bar(x_pos, mean_values, color=cmap2.to_rgba(mean_values))
```

```python
patterns = ('-', '+', 'x', '\\', '*', 'o', 'O', '.')
fig = plt.gca()
mean_value = range(1,len(patterns)+1)
x_pos = list(range(len(mean_value)))
bars = plt.bar(x_pos, mean_value, color='white', edgecolor='black')
for bar,pattern in zip(bars, patterns):
    bar.set_hatch(pattern)
```
